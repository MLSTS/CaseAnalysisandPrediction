## Background

In recent times, [deep neural network](https://en.wikipedia.org/wiki/Deep_learning#Deep_neural_networks) such as [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) has become the standard choice for the complex problem such as [Computer Vision (CV)](https://en.wikipedia.org/wiki/Computer_vision) that allows computers to understand, analyze and labeled images including medical image classifications. This case aims to investigate the presence of [Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) from a collection of chest X-ray image samples. The classification was implemented by constructing a [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) model to extract a significance features from a given chest X-ray image and classify if the patient has a [Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) or not.

[Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) is an inflammatory condition of the lung, affecting primarily the small air sacs known as alveoli, caused by infected with viruses, bacteria or fungi.

The main objective of this case is to try on various [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) models in order to find the efficient [Neural Network](https://en.wikipedia.org/wiki/Artificial_neural_network) model to handle the [Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) classification problem. The results are evaluated and compared using [confusion matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html), [classification report](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html), [ROC curve](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html) and the [Sensitivity at Specificity](https://www.tensorflow.org/api_docs/python/tf/keras/metrics/SensitivityAtSpecificity).

## Model 1: Original data with basic [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) model

In first model, we create the basic [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) model to classify our samples.

### Create Generator
Create generator object for train, validation and test set.

&#10004; Rescaling each pixels RGB value from range [0,255] to range [0,1]. <br>
&#10004; Creating generator object for train, validation and test set which are assigned to the train_generator, valid_generator and test_generator variable. <br>
&#10004; Resizing images from the original size to 150 x 150 pixel. <br>
&#10004; Using 'binary' as class mode in order to classify "NORMAL" or "PNEUMONIA".

```python
train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(
    train_dir, # Path to train directory
    target_size=(150,150),
    batch_size=220, # Choosing batch size by 3512 (Total train samples) / 16 (Step per Epoch)
    shuffle=True, # Shuffle the train set
    class_mode='binary')

validation_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = validation_datagen.flow_from_directory(
    validation_dir, # Path to validation directory
    target_size=(150,150),
    batch_size=147, # Choosing batch size by 1172 (Total validation samples) / 8 (Step per Epoch)
    shuffle=False, # Never shuffle the validation set
    class_mode='binary')

test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir, # Path to test directory
    target_size=(150,150),
    batch_size=147, # Choosing batch size by 11712 (Total test samples) / 8 (Step per Epoch)
    shuffle=False, # Never shuffle the test set
    class_mode='binary')
```

### Create Model

In [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) part we use mutiple [2D Convolution](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2D) layers and [Max Pooling](https://www.tensorflow.org/api_docs/python/tf/keras/layers/MaxPool2D) layers to convert images into the matrics.<br>
Then, we use [Flatten](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Flatten) layer to convert the data from matrics into one column. <br>

After that, we pass the data into neural network with 512 dense. <br>
Finally, we pass the data into 1 dense layer as an output with 'sigmoid' activation (because we want the result in binary). <br>

#### Design Model

```python
# Convolution layers
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten layer
model.add(layers.Flatten())

# Classification layers
model.add(layers.Dense(512, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
          
model.summary()
```

#### Compile Model

```python
M = ['accuracy', FalsePositives(), FalseNegatives(), SensitivityAtSpecificity(0.90)] # Define model metrics
model.compile(loss = 'binary_crossentropy', optimizer = optimizers.RMSprop(lr = 1e-4), metrics = M) # Compile the model
```

#### Train Model

```python
E = 8 # Number of epochs (2**3)

time_start = time.time() # Start the timer

history = model.fit_generator( # Train model
    train_generator,
    steps_per_epoch = 16, # Choose steps per epoch by 220 (Batch size) * 16 = 3512 (Total train sample)
    verbose = 0,
    epochs = E,
    validation_data = validation_generator,
    validation_steps = 8) # Choose validation steps by 147 (Batch size) * 8 = 1172 (Total validation sample)

time_end = time.time() # Stop the timer
time_elapsed = time_end - time_start # Calculate the fitting time
time_per_epoch = time_elapsed / E # Calculate the training time per epoch

print("Model trained")
print(f"Time elapsed: {time_elapsed:.0f} seconds")
print(f"Time per epoch: {time_per_epoch:.2f} seconds")

model.save('Case2_Model_1.h5') # Save the model
print("Model saved")
```

## Model 2: Augmented data with basic [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) model

In second model, we use the same model as the model 1 but we use technique called Image data augmentation to increase the sample size and reduce overfitting.

Image data augmentation is use to manipulate the limited sample. For example, if we have 10 images as a train data. We can use augmentation by flip, rotate, zoom or shift in order to create new images for the training set. This augmentation also reduce overfiting because model train from vary view point and model have more data for training.

### Create Generator

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest'
                                  )
train_generator = train_datagen.flow_from_directory(
    train_dir, # Path to train directory
    target_size=(150,150),
    batch_size=220, # Choosing batch size by 3512 (Total train samples) / 16 (Step per Epoch)
    shuffle=True, # Shuffle the train set
    class_mode='binary')
```

#### Design Model

We use the same mode architecture as the model 1.

```python
# Convolution layers
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten layer
model.add(layers.Flatten())

# Classification layers
model.add(layers.Dense(512, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
          
model.summary()
```

#### Compile Model

```python
M = ['accuracy', FalsePositives(), FalseNegatives(), SensitivityAtSpecificity(0.90)] # Define model metrics
model.compile(loss = 'binary_crossentropy', optimizer = optimizers.RMSprop(lr = 1e-4), metrics = M)
```

#### Train Model

```python
E = 8 # Number of epochs (2**3)

time_start = time.time() # Start the timer

history = model.fit_generator( # Train model
    train_generator,
    steps_per_epoch = 16, # Choose steps per epoch by 220 (Batch size) * 16 = 3512 (Total train sample)
    verbose = 0,
    epochs = E,
    validation_data = validation_generator,
    validation_steps = 8) # Choose validation steps by 147 (Batch size) * 8 = 1172 (Total validation sample)

time_end = time.time() # Stop the timer
time_elapsed = time_end - time_start # Calculate the fitting time
time_per_epoch = time_elapsed / E # Calculate the training time per epoch

print("Model trained")
print(f"Time elapsed: {time_elapsed:.0f} seconds")
print(f"Time per epoch: {time_per_epoch:.2f} seconds")

model.save('Case2_Model_2.h5') # Save the model
print("Model saved")
```

## Model 3: Augmented data with Pretrained ConvNet [(VGG16)](https://keras.io/applications/#vgg16)

In third model, we use pretrained [(VGG16)](https://keras.io/applications/#vgg16) model to classify our samples. [(VGG16)](https://keras.io/applications/#vgg16) is a [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) for Large-Scale Image Recognition. The model has a dataset of over 14 million images belonging to 1000 classes and the most unique thing about [(VGG16)](https://keras.io/applications/#vgg16) is they focused on having convolution layers of 3x3 filter with a stride 1 and always used same padding and maxpool layer of 2x2 filter of stride 2 (you will see in model summary).

### Create Generator

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest'
                                  )
train_generator = train_datagen.flow_from_directory(
    train_dir, # Path to train directory
    target_size=(150,150),
    batch_size=220, # Choosing batch size by 3512 (Total train samples) / 16 (Step per Epoch)
    shuffle=False, # Not shuffle the train set in VGG16
    class_mode='binary')

test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir, # Path to test directory
    target_size=(150,150),
    batch_size=147, # Choosing batch size by 11712 (Total test samples) / 8 (Step per Epoch)
    shuffle=False, # Never shuffle the test set
    class_mode='binary')
```

### Create Model

```python
from tensorflow.keras.applications import VGG16 # Import VGG16 model
conv_base = VGG16(weights='imagenet', include_top=False, input_shape=(150,150,3)) # Create VGG16 model
conv_base.trainable = False
conv_base.summary()
```

#### Extract Sample Feature

```python
train_labels = train_generator.classes # Get labels from the train generator
train_generator.class_mode = False
train_feature_batch = conv_base.predict(train_generator) # Extract feature from the train set using VGG16
train_feature_batch.shape
```

```python
test_labels = test_generator.classes # Get labels from the test generator
test_generator.class_mode = False
test_feature_batch = conv_base.predict(test_generator) # Extract feature from the test set using VGG16
test_feature_batch.shape
```

#### Design Model

Create another model to flatten the result from [VGG16](https://keras.io/applications/#vgg16) and pass data into the [Neural Network](https://en.wikipedia.org/wiki/Artificial_neural_network) for classification.

```python
model = models.Sequential()

# Flatten layer
model.add(tf.keras.layers.Flatten(input_shape=(4,4,512)))

# Classification layer
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
          
model.summary()
```

#### Compile Model

```python
M = ['accuracy', FalsePositives(), FalseNegatives(), SensitivityAtSpecificity(0.90)] # Define model metrics
model.compile(loss = 'binary_crossentropy', optimizer = optimizers.RMSprop(lr = 1e-4), metrics = M)
```

#### Train Model

In this fitting part, we do not have validation data set because model below is just ending part of <br>
hole network(CNN+NW).This fitting code below just represent fitting data into last 3 layers model that we just<br>
created after we fitting VGG16.So validation part is in VGG16 model.

```python
E = 8 # Number of epochs (2**3)

time_start = time.time() # Start the timer

history = model.fit(train_feature_batch, # Train model
                    train_labels,
                    steps_per_epoch=8,
                    verbose=0,
                    epochs=E,
                    use_multiprocessing=True
                   )

time_end = time.time() # Stop the timer
time_elapsed = time_end - time_start # Calculate the training time
time_per_epoch = time_elapsed / E # Calculate the training time per epoch

print("Model trained")
print(f"Time elapsed: {time_elapsed:.0f} seconds")
print(f"Time per epoch: {time_per_epoch:.2f} seconds")

model.save('Case2_Model_3.h5') # Save the model
print("Model saved")
```
