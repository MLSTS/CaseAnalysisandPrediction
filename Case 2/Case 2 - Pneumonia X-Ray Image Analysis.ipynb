## Background

In recent times, [deep neural network](https://en.wikipedia.org/wiki/Deep_learning#Deep_neural_networks) such as [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) has become the standard choice for the complex problem such as [Computer Vision (CV)](https://en.wikipedia.org/wiki/Computer_vision) that allows computers to understand, analyze and labeled images including medical image classifications. This case aims to investigate the presence of [Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) from a collection of chest X-ray image samples. The classification was implemented by constructing a [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) model to extract a significance features from a given chest X-ray image and classify if the patient has a [Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) or not.

[Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) is an inflammatory condition of the lung, affecting primarily the small air sacs known as alveoli, caused by infected with viruses, bacteria or fungi.

The main objective of this case is to try on various [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) models in order to find the efficient [Neural Network](https://en.wikipedia.org/wiki/Artificial_neural_network) model to handle the [Pneumonia](https://en.wikipedia.org/wiki/Pneumonia) classification problem. The results are evaluated and compared using [confusion matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html), [classification report](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html), [ROC curve](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html) and the [Sensitivity at Specificity](https://www.tensorflow.org/api_docs/python/tf/keras/metrics/SensitivityAtSpecificity).

## Preparation

### Import Library
   
First, we import the librarys that we are going to use.

```python
%pylab inline
import time # Timing
import random # Random generators
import pandas as pd # Pandas dataframe
import tensorflow as tf # Tensorflow
from tensorflow.keras import layers, models, optimizers # Neural Network
from tensorflow.keras.preprocessing.image import ImageDataGenerator # Image Preprocessing
from tensorflow.keras.metrics import Accuracy, FalsePositives, FalseNegatives, SensitivityAtSpecificity # Metrics
from sklearn.metrics import classification_report, confusion_matrix, roc_curve # Report
random.seed(10) # Set seed for the random generators
print(f"Tensorflow version: {tf.__version__}")
```

### Data Preprocessing

The dataset we are using in this case is already preprocessing but we omitted the code to reduce the length of this case and provided the step-by-step instruction instead.

Step-by-Step: <br>
&#10004;  Read data from the train directory and assigned the train_normal_images variable for the NORMAL and train_pneumonia_images for the PNEUMONIA. <br>
&#10004;  Read data from the test directory and assigned the test_normal_images variable for the NORMAL and test_pneumonia_images for the PNEUMONIA. <br>

```python
train_dir = 'Case2_ChestX-Ray/train/' # Path to train directory
validation_dir = 'Case2_ChestX-Ray/validation/' # Path to validation directory
test_dir = 'Case2_ChestX-Ray/test/' # Path to test directory
```

```python
result = {} # Create empty dictionary for storing the result from each model
```

### Create Generator
Create generator object for train, validation and test set.

&#10004; Rescaling each pixels RGB value from range [0,255] to range [0,1]. <br>
&#10004; Augmenting images with various technique. <br>
&#10004; Creating generator object for train, validation and test set which are assigned to the train_generator, valid_generator and test_generator variable. <br>
&#10004; Resizing images from the original size to 150 x 150 pixel. <br>
&#10004; Using 'binary' as class mode in order to classify "NORMAL" or "PNEUMONIA".

Image data augmentation is use to manipulate the limited sample. For example, if we have 10 images as a train data. We can use augmentation by flip, rotate, zoom or shift in order to create new images for the training set. This augmentation also reduce overfiting because model train from vary view point and model have more data for training.

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=10,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest'
                                  )
train_generator = train_datagen.flow_from_directory(
    train_dir, # Path to train directory
    target_size=(150,150),
    batch_size=220, # Choosing batch size by 3512 (Total train samples) / 16 (Step per Epoch)
    shuffle=True, # Shuffle the train set
    seed=10,
    class_mode='binary')

validation_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = validation_datagen.flow_from_directory(
    validation_dir, # Path to validation directory
    target_size=(150,150),
    batch_size=147, # Choosing batch size by 1172 (Total validation samples) / 8 (Step per Epoch)
    shuffle=False, # Never shuffle the validation set
    class_mode='binary')

test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir, # Path to test directory
    target_size=(150,150),
    batch_size=147, # Choosing batch size by 11712 (Total test samples) / 8 (Step per Epoch)
    shuffle=False, # Never shuffle the test set
    class_mode='binary')
```

## Model 1: Augmented data with basic [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) model

In first model, we create the basic [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) model to classify our samples.

### Create Model

In [Convolutional Neural Network (CNN)](https://en.wikipedia.org/wiki/Convolutional_neural_network) part we use mutiple [2D Convolution](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2D) layers and [Max Pooling](https://www.tensorflow.org/api_docs/python/tf/keras/layers/MaxPool2D) layers to convert images into the matrics.<br>
Then, we use [Flatten](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Flatten) layer to convert the data from matrics into one column. <br>

After that, we pass the data into neural network with 512 dense. <br>
Finally, we pass the data into 1 dense layer as an output with 'sigmoid' activation (because we want the result in binary). <br>

#### Design Model

```python
model = models.Sequential()

# Convolution layers
model.add(layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten layer
model.add(layers.Flatten())

# Classification layers
model.add(layers.Dense(512, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
          
model.summary()
```

#### Compile Model

```python
M = ['accuracy', FalsePositives(), FalseNegatives(), SensitivityAtSpecificity(0.90)] # Define model metrics
model.compile(loss = 'binary_crossentropy', optimizer = optimizers.RMSprop(lr = 1e-4), metrics = M) # Compile the model
```

#### Train Model

```python
E = 8 # Number of epochs (2**3)

time_start = time.time() # Start the timer

history = model.fit_generator( # Train model
    train_generator,
    steps_per_epoch = 16, # Choose steps per epoch by 220 (Batch size) * 16 = 3512 (Total train sample)
    verbose = 0,
    epochs = E,
    validation_data = validation_generator,
    validation_steps = 8) # Choose validation steps by 147 (Batch size) * 8 = 1172 (Total validation sample)

time_end = time.time() # Stop the timer
time_elapsed = time_end - time_start # Calculate the fitting time
time_per_epoch = time_elapsed / E # Calculate the training time per epoch

print("Model trained")
print(f"Time elapsed: {time_elapsed:.0f} seconds")
print(f"Time per epoch: {time_per_epoch:.2f} seconds")

model.save('Case2_Model_1.h5') # Save the model
print("Model saved")
```

#### Analyze Model

```python
history.history.keys()
```

```python
# Assign new variables for analyze
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
false_positives = history.history['false_positives']
val_false_positives = history.history['val_false_positives']
false_negatives = history.history['false_negatives']
val_false_negatives = history.history['val_false_negatives']
sensitivity_at_specificity = history.history['sensitivity_at_specificity']
val_sensitivity_at_specificity = history.history['val_sensitivity_at_specificity']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))
```

```python
fig, axs = subplots(2, 2, figsize=(15,8))

# Plot training and validation accuracy
axs[0][0].plot(epochs, accuracy, 'bo-', label='Training accuracy')
axs[0][0].plot(epochs, val_accuracy, 'r*-', label='Validation accuracy')
axs[0][0].set_title('Training and validation accuracy')
axs[0][0].grid ()
axs[0][0].legend()

# Plot training and validation loss
axs[0][1].plot(epochs, loss, 'bo-', label='Training loss')
axs[0][1].plot(epochs, val_loss, 'r*-', label='Validation loss')
axs[0][1].set_title('Training and validation loss')
axs[0][1].grid()
axs[0][1].legend()
# Plot training and validation sensitivity at specificity
axs[1][0].plot(epochs, sensitivity_at_specificity, 'bo-', label='Training sensitivity at specificity')
axs[1][0].plot(epochs, val_sensitivity_at_specificity, 'r*-', label='Validation sensitivity at specificity')
axs[1][0].set_title('Training and validation sensitivity at specificity')
axs[1][0].grid ()
axs[1][0].legend()

# Plot validation prediction and validation label
validation_labels = validation_generator.classes
validation_predictions = model.predict_generator(validation_generator).flatten()
axs[1][1].plot(validation_predictions, label = 'Prediction', color = 'g')
axs[1][1].plot(validation_labels, label = 'Label', color = 'r')
axs[1][1].set_title('Validation prediction and Validation label')
axs[1][1].grid()
axs[1][1].legend()

show()
```

```python
# Predict the test set and compute prediction time.
start_time = time.time()
predictions = model.predict_generator(test_generator)
print(f"Predicting time: {(time.time() - start_time):.0f} seconds")
```

```python
labels = test_generator.classes # Get labels from the test generator
reshaped_predictions = predictions.reshape(predictions.shape[0]) # Reshape predictions to be 1 dimension array
```

```python
fig, axs = subplots(1, 2, figsize=(10,4))

# plot ROC Curve
fpr, tpr, threshold = roc_curve(labels, reshaped_predictions)
axs[0].plot(fpr, tpr, 'b')
axs[0].plot([0, 1], [0, 1],'r--')
axs[0].set_title('Receiver Operating Characteristic')
axs[0].set_xlim([0, 1])
axs[0].set_ylim([0, 1])
axs[0].set_ylabel('True Positive Rate')
axs[0].set_xlabel('False Positive Rate')
axs[0].grid()

#Find the best threshold
axs[1].plot(threshold, 1 - fpr, label = 'Specificity')
axs[1].plot(threshold, tpr, label = 'Sensitivity')
axs[1].axvline(0.5, color = 'red', linestyle = ':')
axs[1].set_title('Threshold value for prediction')
axs[1].set_xlim([0, 1])
axs[1].set_xlabel('Threshold')
axs[1].set_ylabel('Metrics value')
axs[1].grid()
axs[1].legend()

show()
```

```python
reshaped_predictions = 1.0 * (predictions > 0.63).reshape(predictions.shape[0])
```

```python
#Show confusion matrix and some other metrics
cm = confusion_matrix(labels, reshaped_predictions)

print(cm)
print(classification_report(labels, reshaped_predictions))

metraic_1 = SensitivityAtSpecificity(0.9)
metraic_1.update_state(labels, predictions.reshape(predictions.shape[0]))
print(f"Sensitivity at 90% Specificity = {metraic_1.result().numpy()}")

sensitivity = cm[1,1] / (cm[1,0] + cm[1,1])
print(f"Sensitivity (Recall) = {sensitivity}")
specificity = cm[0,0] / (cm[0,0] + cm[0,1])
print(f"Specificity = {specificity}")

metraic_2 = FalseNegatives()
metraic_2.update_state(labels, reshaped_predictions.reshape(predictions.shape[0]))
falsenegative = metraic_2.result().numpy()
print(f"False negative = {falsenegative}")

metraic_3 = history.history['accuracy'][len(history.history['accuracy'])-1]
accuracy = metraic_3
print(f"Accuracy = {accuracy}")

# Save result in the result dictionary
result["Model 1"] = {"Sensitivity":sensitivity,
                     "Specificity":specificity,
                     "False Negative":falsenegative,
                     "Accuracy":accuracy,
                     "Training Time":time_elapsed
                    }
```

## Model 2: Augmented data with increase kernel size from the model 1

## Model 3: Augmented data with mimic classical [LeNet-5](https://engmrk.com/lenet-5-a-classic-cnn-architecture/) architecture model

Yann LeCun, Leon Bottou, Yosuha Bengio and Patrick Haffner proposed a neural network architecture for handwritten and machine-printed character recognition in 1990’s which they called LeNet-5. The architecture is straightforward and simple to understand that’s why it is mostly used as a first step for teaching Convolutional Neural Network. The LeNet-5 architecture consists of two sets of convolutional and average pooling layers, followed by a flattening convolutional layer, then two fully-connected layers and finally a softmax classifier. - [Source](https://engmrk.com/lenet-5-a-classic-cnn-architecture/)

### Create Model

#### Design Model

```python
model = models.Sequential()
model.add(layers.Conv2D(6, (5,5), activation='relu', input_shape=(150,150, 3)))
model.add(layers.MaxPooling2D((2,2), strides=2))
model.add(layers.Conv2D(16, (5,5), activation = 'relu'))
model.add(layers.MaxPooling2D((2,2), strides=2))
model.add(layers.Conv2D(120, (5,5), activation = 'relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(84, activation ='relu'))
model.add(layers.Dropout(0.4))
model.add(layers.Dense(1, activation = 'sigmoid'))

model.summary()
```

#### Compile Model

```python
M = ['accuracy', FalsePositives(), FalseNegatives(), SensitivityAtSpecificity(0.90)] # Define model metrics
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = M)
```

#### Train Model

```python
E = 8 # Number of epochs (2**3)

time_start = time.time() # Start the timer

history = model.fit_generator( # Train model
    train_generator,
    steps_per_epoch = 16, # Choose steps per epoch by 220 (Batch size) * 16 = 3512 (Total train sample)
    verbose = 0,
    epochs = E,
    validation_data = validation_generator,
    validation_steps = 8) # Choose validation steps by 147 (Batch size) * 8 = 1172 (Total validation sample)

time_end = time.time() # Stop the timer
time_elapsed = time_end - time_start # Calculate the fitting time
time_per_epoch = time_elapsed / E # Calculate the training time per epoch

print("Model trained")
print(f"Time elapsed: {time_elapsed:.0f} seconds")
print(f"Time per epoch: {time_per_epoch:.2f} seconds")

model.save('Case2_Model_2.h5') # Save the model
print("Model saved")
```
